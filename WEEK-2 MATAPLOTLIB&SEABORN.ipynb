```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("tips.csv")# read the dataset
df.head()
```

```python
x=df['day']# bar graph
y=df['total_bill']
plt.title("Tips Dataset")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.bar(x,y,color='blue')#plt.barh(x,y,color='blue') plot,scatter
plt.show()
```

```python
plt.barh(x,y,color='blue')
```

```python
plt.plot(x,y,color='blue')
```

```python
plt.scatter(x,y,color='blue')
```

```python
#seaborn
import seaborn as sn
df1=pd.read_csv("nba.csv")
df1.head()
```

```python
#barplot
ax=sn.barplot(x='Age',y='Weight',data=df1)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.tight_layout()
plt.show()
```

```python
#line plot
sn.lineplot(x='Age',y='Weight',hue="Position",data=df1)
plt.tight_layout()
plt.show()
```
